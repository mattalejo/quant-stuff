```python
from fastquant import get_pse_data
```

```python
from matplotlib import pyplot as plt
```

```python
df_train = get_pse_data("JFC", "2005-01-01", "2018-12-31")
df_train['close'].plot(figsize=(10, 6))
plt.title("Daily Closing Prices of JFC\nfrom 2018-01-01 to 2019-01-01", fontsize=20)
```

```python
returns = df_train['close'].pct_change()
returns.plot(figsize=(10, 6))
plt.title("Daily Returns of JFC\nfrom 2018-01-01 to 2019-01-01", fontsize=20)
```

```python
ax1 = plt.figure().add_axes([0.1,0.1,0.8,0.8])
returns.plot.hist(bins=60)
```

```python
print(returns)
```

```python
print(returns.mean())
```

```python
print(returns.std())
```
